```python
# ライブラリのインポート
import pandas as pd
from datetime import datetime as dt

# CSVファイル読込、日付型項目を索引、時系列データ集計(parse_dates)
frame1 = pd.read_csv('C:\\Users\\Owner\\Pandas-Exercises7\FILE.csv', index_col=['取引日'], parse_dates=True)
```

```python
# 時系列データの週単位集計(平均値)の最初を表示
frame1.resample(rule='W').mean().head()
```

```python
# 時系列データの週単位集計(合計値)の最初を表示
frame1.resample(rule='W').sum().head()
```
